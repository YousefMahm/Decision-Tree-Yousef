```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
data = {
'Income': [50000, 30000, 70000, 20000, 90000],
'Age': [25, 35, 40, 22, 30],
'Approved': ['Yes', 'No', 'Yes', 'No', 'No']
}
df = pd.DataFrame(data)


# Features and target
X = df[['Income', 'Age']]
y = df['Approved']
```

```python
clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
clf = clf.fit(X, y)
```

```python
plt.figure(figsize=(10,6))
tree.plot_tree(clf, feature_names=['Income', 'Age'], class_names=['No','Yes'], filled=True, fontsize=10)
plt.show()
```

```python
# Predict new applicants
new_applicants = pd.DataFrame({
'Income': [60000, 25000],
'Age': [28, 33]
})


predictions = clf.predict(new_applicants)
print(predictions)
# Output: ['Yes' 'No']
```
